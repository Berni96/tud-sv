- __Signalrepräsentation__

```python
'''
Beispiel: Exponentielles Signal mit Realzahlen
'''

# Package importieren
import numpy as np
import matplotlib.pyplot as plt

# Zeitbereich
t = np.linspace(-5, 5, 100)    

# Exponentielle Funktion
s = np.exp(0.7*t)            

# Plot
plt.gcf().set_size_inches(8, 5)  # Größe setzen
plt.title(u's = exp(0.7*t)')  # Titel setzen
plt.plot(t, s)   
plt.show()  # Plot zeigen
```

```python
'''
Beispiel: Exponentielles Signal mit komplexen Zahlen
'''
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-5, 5, 1000)
s = np.exp((complex(-0.7, 7))*t)

plt.subplot(221)
plt.title(u'Real')
plt.plot(t,np.real(s))
plt.subplot(222)
plt.title(u'Image')
plt.plot(t,np.imag(s))
plt.subplot(223)
plt.title(u'Amplitude')
plt.plot(t,np.abs(s))
plt.subplot(224)
plt.title(u'Phase')
plt.plot(t,np.angle(s))

plt.gcf().set_size_inches(15, 10)
plt.show()
```

```python
'''
Aufgabe 1.1: Sinus- bzw. Cosinussignal mit verschiedenen Frequenzen impliementieren
w = pi/3, pi, 2pi/3, 2pi
'''
import numpy as np
import matplotlib.pyplot as plt

# Zeitbereich
t = np.linspace(0, 10, 1000)

# w = pi/3, Sinus:
s1 = np.sin(np.pi/3 * t)
plt.subplot(221)
plt.plot(t, s1)
plt.title(u'sin((pi/3)*t)')

# w = pi, Sinus:
s2 = np.sin(np.pi * t)
plt.subplot(222)
plt.plot(t, s2)
plt.title(u'sin(pi*t)')

# w = 2pi/3, Cosinus:
s3 = np.cos(2/3 * np.pi * t)
plt.subplot(223)
plt.plot(t, s3)
plt.title(u'cos((2pi/3)*t)')

# w = 2pi, Cosinus:
s4 = np.cos(2 * np.pi * t)
plt.subplot(224)
plt.plot(t, s4)
plt.title(u'cos((2pi)*t)')

plt.gcf().set_size_inches(15, 10)
plt.show()
```

```python
'''
Aufgabe 1.2: Impulssignal impliementieren
'''
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-5, 5, 1000) 

# Impulsbreite = 3
plt.subplot(121)
s1 = np.where((t >= -2) & (t <= 1), 1, 0) 
plt.title(u'Impulsbreite = 3')
plt.plot(t, s1)

# Impulsbreite = 0.1
plt.subplot(122)
s2 = np.where((t >= -0.05) & (t <= 0.05), 1, 0) 
plt.title(u'Impulsbreite = 0.1')
plt.plot(t, s2)

plt.gcf().set_size_inches(15, 5)
plt.show()
```

```python
'''
Aufgabe 1.3: Sprungsignal impliementieren
'''
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-5, 5, 1000) 
s = np.where(t >= -2, 1, 0) 
plt.plot(t, s)

plt.gcf().set_size_inches(8, 5)
plt.show()
```

```python
'''
Beispiel: Diskrete Exponentielle Signale
'''
import numpy as np
import matplotlib.pyplot as plt

# Anzahl von Samplen
n = np.arange(0, 20)

# Diskretes Signal
s = np.exp(-0.3*n)

# Wertebereich zum Anzeigen
plt.ylim(-0.5, 1.5)  

# Stem plot
plt.stem(n, s, use_line_collection=True)

plt.gcf().set_size_inches(8, 5)
plt.show()
```

```python
'''
Aufgabe 1.4: Diskrete Sinus- und Cosinussignale implementieren
'''
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(0, 50)

plt.subplot(221)
plt.title(u'sin(npi)')
plt.stem(n, np.sin(n*np.pi), use_line_collection=True)
plt.subplot(222)
plt.title(u'sin(npi/6)')
plt.stem(n, np.sin(n*np.pi/6), use_line_collection=True)
plt.subplot(223)
plt.title(u'cos(npi/9)')
plt.stem(n, np.cos(n*np.pi/9), use_line_collection=True)
plt.subplot(224)
plt.title(u'cos(2n)')
plt.stem(n,np.cos(2*n), use_line_collection=True)

plt.gcf().set_size_inches(15, 10)
plt.show()
```

- __Operationen von Signalen__

```python
'''
Beispiel: Addition und Multiplikation von Signalen
'''
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-5, 5, 1000)
s1 = np.sin(2*np.pi*t)
s2 = np.exp(-t/2)

plt.subplot(221)
plt.title(u's1')
plt.plot(t, s1)
plt.subplot(222)
plt.title(u's2')
plt.plot(t, s2)
plt.subplot(223)
plt.title(u's1+s2')
plt.plot(t, s1+s2)
plt.subplot(224)
plt.title(u's1xs2')
plt.plot(t, s1*s2)

plt.gcf().set_size_inches(15, 10)
plt.show()
```

```python
'''
Aufgabe 1.7: Addition und Multiplikation von diskreten Signalen
'''
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-5, 5)
s1 = np.cos(5*n)
s2 = np.exp(-0.3*n)

plt.subplot(221)
plt.title(u's1')
plt.stem(n, s1, use_line_collection=True)
plt.subplot(222)
plt.title(u's2')
plt.stem(n, s2, use_line_collection=True)
plt.subplot(223)
plt.title(u's1+s2')
plt.stem(n, s1+s2, use_line_collection=True)
plt.subplot(224)
plt.title(u's1xs2')
plt.stem(n, s1*s2, use_line_collection=True)

plt.gcf().set_size_inches(15, 10)
plt.show()
```

```python
'''
Beispiel: Faltung von kontinuierlichen Signalen
'''
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-5, 5, 1000)
s1 = np.where((t >= -2) & (t <= 3), 1, 0) 
s2 = np.where(t <= 4, np.exp(0.7*t), 0)

# Faltung
s = np.convolve(s1, s2)
ts = np.linspace(-10, 9, len(s))

plt.subplot(221)
plt.title(u's1')
plt.plot(t, s1)
plt.subplot(222)
plt.title(u's2')
plt.plot(t, s2)
plt.subplot(212)
plt.title(u's1*s2')
plt.plot(ts, s)

plt.gcf().set_size_inches(10, 10)
plt.show()
```

```python
'''
Aufgabe 1.8: Faltung von diskreten Signalen
'''
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-5, 5)
s1 = np.where((n >= -2) & (n <= 3), 1, 0) 
s2 = np.where(n >= -3, np.exp(-0.7*n), 0)

s = np.convolve(s1, s2)
ns = range(-10, 9)

plt.subplot(221)
plt.title(u's1')
plt.stem(n, s1, use_line_collection=True)
plt.subplot(222)
plt.title(u's2')
plt.stem(n, s2, use_line_collection=True)
plt.subplot(212)
plt.title(u's1*s2')
plt.stem(ns, s, use_line_collection=True)

plt.gcf().set_size_inches(10, 10)
plt.show()
```

```python
'''
Aufgabe 2: Fourier-Transformation mittels FFT
(kontinuierliches und nichtperiodisches Signal -> kontinuierliches und nichtperiodisches Spektrum)
'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft

t = np.linspace(-2*np.pi, 2*np.pi, 200)  
s = np.sinc(np.pi * t)  
S = fft(s)
st = ifft(S)

plt.subplot(311)
plt.title(u'Zeitbereich')
plt.plot(t, s)
plt.subplot(312)
plt.title(u'Amplitutsgang')
plt.plot(np.abs(np.fft.fftshift(S)))
plt.subplot(313)
plt.title(u'Inverse Fourier-Transformation')
plt.plot(t, st)
             
plt.gcf().set_size_inches(10, 15)
plt.show()      
```

__\* Bemerkung__: Rechner können bloß diskrete Signal verarbeiten, d.h. die Fourier-Transformation von kontinuierlichen Signalen bzw. Spektralen von Computer nicht realisierbar.
